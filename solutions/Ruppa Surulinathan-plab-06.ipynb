### c) Using generators to read entries from FASTA files

```python
def fasta_sequences(file):
        name, seq = None, []
        for line in file:
            line = line.rstrip()
            if line.startswith(">"):
                if name: yield (name, ''.join(seq))
                name, seq = line, []
            else:
                seq.append(line)
        if name: yield (name, ''.join(seq))
```

```python
f  =  open("ecoli-genes.ffn")
fasta_seq = fasta_sequences(f)
next(fasta_seq)
```

### Ex.2  Complementary DNA

```python

def finding_complement(sequence):
    complement = {'A': 'T', 'C': 'G', 'G': 'C', 'T': 'A'}
    sequence_comp = "".join(complement.get(base,'N') for base in sequence)
    return sequence_comp[::-1]
```

### Ex.3 Finding genes in prokaryotes: OpenReadingFrames

```python
def find_ORF(sequence):
    
     
    readingframe_list = []
        
    for readingframe_n in range(0,3):
        
        readingframe = ''
        start_pos = None
        
        for i in range(readingframe_n, len(sequence), 3):
            codon = sequence[i:i+3]
            if codon == "ATG" and start_pos == None:
                readingframe = readingframe + codon
                start_pos = i
                continue
            
            if start_pos != None:
                if codon not in ["TAA", "TAG", "TGA"]:
                    readingframe = readingframe + codon
                else: 
                    readingframe = readingframe + codon
                    readingframe_list.append((readingframe, start_pos+1, i+3)) 
                    readingframe = ''
                    start_pos = None
                    
        
    return readingframe_list

    
```
