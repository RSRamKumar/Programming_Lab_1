# Ex.2 The Dotplot matrix

## (a) Generate dotplot matrix

```python

import numpy as np
def create_mat(len_seq1,len_seq2):
    """function to return zero filled matrix based on the input lengths
    Arg 1: length of sequence 1
    Arg 2: length of sequence 2
    Return : zero filled matrix"""
    dp=np.zeros((len_seq1,len_seq2),dtype=int)
    return dp

def dotplot (seq1, seq2, window,stringency):
    """function to do dotplot matrix
    Arg 1: sequence 1, Arg 2: sequence 2, Arg 3: window size, Arg 4: stringency value
    Return: dotplot matrix"""
    mat = create_mat(len(seq1),len(seq2))
    start =int(window/2)
    for i in range(start,len(seq1) - start):
        for j in range(start,len(seq2)-start):
            matches = 0
            l=j-start
            for k in range(i-start, i+start+1):
                if seq1[k] == seq2[l]: matches  += 1
                l+=1 
                if matches >= stringency: mat[i][j] = 1
    return mat


def print_dotplot(mat, s1, s2):
    """function to do print dotplot in the screen
    Arg 1: dotplot matrix , Arg 2: sequence 1, Arg 3: sequence 2
    Return: dotplot"""
    sys.stdout.write(" "+"|" + s2+"\n")
    for i in range( len (mat)):
        sys.stdout.write(s1[i]+"|")
        for j in range( len (mat[i])):
            if mat[i][j] == 1:
                sys.stdout.write("∗")
            else :
                sys.stdout.write(".")
        sys.stdout.write("\n")

s1 = "MAXISTAYAWAYATSIXAM"
s2 = "MAXISTAYAWAYATSIXAM"
matrix =dotplot(s1, s2, 1,1)
print("Dotplot Matrix for the given sequences are as follows:\n", matrix)
print("\n")
print("Dotplot Representation is as follows:")
print_dotplot(matrix, s1, s2)
```

## (b) The Dotplot as ASCII art

```python

def dotplot2Ascii(dp,seqA,seqB,heading,filepath):
    """function to do write dotplot in the given .txt file
    Arg 1: dotplot matrix , Arg 2: sequence 1, Arg 3: sequence 2, Arg 4: heading, Arg 5: output file name
    Return: dotplot in the given .txt file"""
    f=open(filepath,'w')
    f.write("\t"+heading+"\n")
    f.write(" "+"|" + seqB+"\n")
    for i in range( len (dp)):
        f.write(seqA[i]+"|")
        for j in range( len (dp[i])):
            if dp[i][j] == 1:
                f.write("+")
            else :
                f.write(".")
        f.write("\n")
    print("The output is written in the file succesfully.")
```

## (c) Graphical output using matplotlib

Using **imshow method**

```python
def dotplot2Graphics_imshow(dp,hdA,hdB,heading,filename):
    """function to plot the dotplot using imshow method in matplotlib
    Arg 1: dotplot matrix, Arg 2:sequence 1, Arg 3: sequence 2, Arg 4: heading, Arg 5: file path
    Return : Image stored in the given file path"""
    plt.figure(figsize = (10,5))
    plt.imshow(dp, interpolation="none",cmap='Paired')
    plt.xlabel("first sequence")
    plt.ylabel("second sequence")
    plt.title(heading)
    
    plt.savefig(filepath+filename)
    plt.show()

    
dp=dotplot(A,B,5,3)
dotplot2Graphics_imshow(dp,A,B,'Dotplot between first and second sequence','imshow.png')
```

```python

def count(S, m, n): 
    """function to calculate the number of possible ways to exchange a coin
    Arg 1: list of coins
    Arg 2: number of coins
    Arg 3: amount to be exchanged
    Return : number of possible solutions"""
    
 
    table = [[0 for x in range(m)] for x in range(n+1)]  
    for i in range(m): 
        table[0][i] = 1

    for i in range(1, n+1): 
        for j in range(m): 
            x = table[i - S[j]][j] if i-S[j] >= 0 else 0
            y = table[i][j-1] if j >= 1 else 0
            table[i][j] = x + y 
    return table[n][m-1] 

coins = [1, 2, 3,4,5] 
m = len(coins) 
amount = 7
print(count(coins, m, amount)) 

```
