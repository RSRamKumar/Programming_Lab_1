```python
# implementing the needleman algorithm
pt ={'match': 1, 'mismatch': -1, 'gap': -8}
def needleman(s1, s2):
    m, n = len(s1),len(s2)
    score = np.zeros((n+1, m+1))
    #Initialization
    
    for i in range(n+1):
        score[i][0] = pt['gap'] * i     # filling columns
        
    for j in range(m+1):
        score[0][j] = pt['gap'] * j     # filling rows
        
    init_x = len(s1)+1
    init_y = len(s2)+1
    for x in range(1,init_x): 
        for y in range(1,init_y):
            
            blosum_score = int(blosum[s2[i-1]+ s1[j-1]]) # x , y
            cond1 = score[y-1,x-1] + blosum_score
            cond2 = score[y,x-1] + pt['gap']
            cond3 = score[y-1,x] + pt['gap']
            value=max(cond1,cond2,cond3)
            score[y,x] = value
    return score

seq_1= 'ELVISISALIVE'
seq_2="PAVLISDEAD"
scoring_mat=pd.DataFrame(needleman(seq_1,seq_2))
scoring_mat.columns=[" "]+[i for i in seq_1]

scoring_mat['index']=[" "]+[i for i in seq_2]
scoring_mat.set_index('index', inplace=True, drop=True)

print(scoring_mat.iloc[-1,-1],"is the alignment score.")
print("The scoring matrix obtained is as follows:")
(scoring_mat)
```

```python
# implementation of global algorithm and traceback
def needleman_wunsch(seq1, seq2):
    
    # Store length of two sequences
    n = len(seq1)  
    m = len(seq2)
    gap_penalty = -1
    match_award = 1
    mismatch_penalty = -8

    # Generate matrix of zeros to store scores
    score = np.zeros((m+1, n+1))
    def match_score(alpha, beta):
        if alpha == beta:
            return match_award
        elif alpha == '-' or beta == '-':
            return gap_penalty
        else:
            return mismatch_penalty

    
   
    # Calculate score table
    
    # Fill out first column
    for i in range(0, m + 1):
        score[i][0] = gap_penalty * i
    
    # Fill out first row
    for j in range(0, n + 1):
        score[0][j] = gap_penalty * j
    
    # Fill out all other values in the score matrix
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            # Calculate the score by checking the top, left, and diagonal cells
            match = score[i - 1][j - 1] + match_score(seq1[j-1], seq2[i-1])
            delete = score[i - 1][j] + gap_penalty
            insert = score[i][j - 1] + gap_penalty
            # Record the maximum score from the three possible scores calculated above
            score[i][j] = max(match, delete, insert )    
    
    # Traceback and compute the alignment 
    
    # Create variables to store alignment
    align1 = ""
    align2 = ""
    
    # Start from the bottom right cell in matrix
    i = m
    j = n
    
    # We'll use i and j to keep track of where we are in the matrix, just like above
    while i > 0 and j > 0: # end touching the top or the left edge
        score_current = score[i][j]
        score_diagonal = score[i-1][j-1]
        score_up = score[i][j-1]
        score_left = score[i-1][j]
        
        # Check to figure out which cell the current score was calculated from,
        # then update i and j to correspond to that cell.
        if score_current == score_diagonal + match_score(seq1[j-1], seq2[i-1]):
            align1 += seq1[j-1]
            align2 += seq2[i-1]
            i -= 1
            j -= 1
        elif score_current == score_up + gap_penalty:
            align1 += seq1[j-1]
            align2 += '-'
            j -= 1
        elif score_current == score_left + gap_penalty:
            align1 += '-'
            align2 += seq2[i-1]
            i -= 1

    # Finish tracing up to the top left cell
    while j > 0:
        align1 += seq1[j-1]
        align2 += '-'
        j -= 1
    while i > 0:
        align1 += '-'
        align2 += seq2[i-1]
        i -= 1
    
    # Since we traversed the score matrix from the bottom right, our two sequences will be reversed.
    # These two lines reverse the order of the characters in each sequence.
    align1 = align1[::-1]
    align2 = align2[::-1]
    
    return(align1, align2)

output1, output2 = needleman_wunsch(seq_1, seq_2)

print(output1 + "\n" + output2)
    
    

```

```python
# reading the blosum as matrix for convenience
def open_blosum(filename = (r"C:\Users\Ram Kumar R S\Downloads\SoSe 2020\blosum62.txt")):
    
    # Read Blosum
    blosum = open(filename).read().splitlines()
    
    blosum_seq = ''.join(blosum[0].split())
    blosum = blosum[1:] # cut header
    blosum = [line[2:] for line in blosum]
    blosum = [line.split() for line in blosum]

    blosum = [list(map(int, line)) for line in blosum[:-1]]
    blosum = pd.DataFrame(blosum)
    
    # Rename Columns
    for i in range(len(blosum_seq)):
        blosum.rename(columns={ blosum.columns[i]: blosum_seq[i] }, inplace = True)

    # Rename Rows
    seq2_dict = defaultdict()
    for i in range(len(blosum_seq)):
        seq2_dict[i] = blosum_seq[i]
    blosum.rename(index=seq2_dict, inplace = True)
    
    
    return blosum

blosum = open_blosum()
blosum
```

# (b) Formatted output

```python

def format_outputting(seq1, seq2, blosum):
    
    # Init
    match_counter = 0
    similarity_counter = 0
    binding = ""  
    
    # Get shorter seq for indexing
    if len(seq1) > len(seq2):
        seq_range = len(seq2)
    else:
        seq_range = len(seq1)


    # Get blosum matrix
    blosum_seq = list(blosum.columns.values)
      

    for i in range(seq_range):
        if seq1[i] == seq2[i]:
            binding = binding + "|"
            match_counter += 1
            
        elif seq1[i] == "-" or seq2[i] == "-":
            binding = binding + " "

        else:
            for j in range(len(blosum_seq)):
                if seq1[i] == blosum_seq[j]:
                    x = j
            for k in range(len(blosum_seq)):
                if seq2[i] == blosum_seq[k]:
                    y = k
            
            if blosum.iloc[x,y] > 0:
                binding = binding + ":"
                similarity_counter += 1
            else:
                binding = binding + " "

    
    # Line break
    seq1 = [ seq1[i:i+80] for i in range(0, len(seq1), 80) ]
    binding = [ binding[i:i+80] for i in range(0, len(binding), 80) ]
    seq2 = [ seq2[i:i+80] for i in range(0, len(seq2), 80) ]
    
    if len(seq1) > len(seq2):
        n_lines = len(seq1)
    else:
        n_lines = len(seq2)
    
    # Visualized alignments
    printing_output = ""
    for i in range(n_lines):
        printing_output += seq1[i] + "\n" + binding[i] + "\n" + seq2[i] + "\n\n\n"
    
    
    sequence_identity = match_counter / seq_range * 100 # matching residues relative to the length of the aligned sequences including gaps
    sequence_similarity = similarity_counter / seq_range * 100 # is the percentage of matched similar residues (i.e.,those with a positive value in the scoring matrix)  relative to the length of the aligned sequences including gaps
    
    return printing_output, sequence_identity, sequence_similarity
        

sequences_visual_alignment, sequence_identity, sequence_similarity = format_outputting(seq_1,seq_2, blosum)
print(sequences_visual_alignment)
```

# (d) Smith-Waterman

```python
pt ={'match': 2, 'mismatch': -1, 'gap': -8}

def mch(alpha, beta):
    if alpha == beta:
        return pt['match']
    elif alpha == '-' or beta == '-':
        return pt['gap']
    else:
        return pt['mismatch']

def swater(s1, s2):
    m, n = len(s1), len(s2)
    H = np.zeros((m+1, n+1))    
    T = np.zeros((m+1, n+1))
    max_score = 0
    # Score, Pointer Matrix
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            sc_diag = H[i-1][j-1] + mch(s1[i-1], s2[j-1])
            sc_up = H[i][j-1] + pt['gap']
            sc_left = H[i-1][j] + pt['gap']
            H[i][j] = max(0,sc_left, sc_up, sc_diag)
            if H[i][j] == 0: T[i][j] = 0
            if H[i][j] == sc_left: T[i][j] = 1
            if H[i][j] == sc_up: T[i][j] = 2
            if H[i][j] == sc_diag: T[i][j] = 3
            if H[i][j] >= max_score:
                max_i = i
                max_j = j
                max_score = H[i][j];
    
    align1, align2 = '', ''
    i,j = max_i,max_j
    
    #Traceback
    while T[i][j] != 0:
        if T[i][j] == 3:
            a1 = s1[i-1]
            a2 = s2[j-1]
            i -= 1
            j -= 1
        elif T[i][j] == 2:
            a1 = '-'
            a2 = s2[j-1]
            j -= 1
        elif T[i][j] == 1:
            a1 = s1[i-1]
            a2 = '-'
            i -= 1
        align1 += a1
        align2 += a2

    align1 = align1[::-1]
    align2 = align2[::-1]
    sym = ''
    iden = 0
    for i in range(len(align1)):
        a1 = align1[i]
        a2 = align2[i]
        if a1 == a2:                
            sym +="|"
            iden += 1
        elif a1 != a2 and a1 != '-' and a2 != '-': 
            sym += ' '
        elif a1 == '-' or a2 == '-':          
            sym += ' '
    
    identity = iden / len(align1) * 100
    print('Identity = %f percent' % identity)
    print('Score =', max_score)
    
    return align1,sym,align2

```

```python
def formatting(seq_1,seq_2):
    align1,sym,align2=swater(seq_1,seq_2)
    seq1 = [ align1[i:i+80] for i in range(0, len(align1), 80) ]
    binding = [ sym[i:i+80] for i in range(0, len(sym), 80) ]
    seq2 = [ align2[i:i+80] for i in range(0, len(align2), 80) ]
    if len(seq1) > len(seq2):
        n_lines = len(seq1)
    else:
        n_lines = len(seq2)
    

    format_output = ""
    for i in range(n_lines):
        format_output += seq1[i] + "\n" + binding[i] + "\n" + seq2[i] + "\n\n"
    return format_output
    
print(formatting(seq_1,seq_2))
```
